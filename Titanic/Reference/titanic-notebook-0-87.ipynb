# 資料科學框架

1. **問題是什麼？**  
   如果資料科學、大數據、機器學習、預測分析、商業智慧，或任何熱門 buzzword 是「解方」，  
   那麼「問題」是什麼？  
   老話一句：**不要本末倒置。**  
   問題 → 需求 → 解決方案 → 設計 → 技術。  
   我們常常太快跳到新的科技、工具或演算法，卻還沒搞清楚我們要解的是什麼問題。

2. **資料集在哪裡？**  
   John Naisbitt 在 1984 年的《Megatrends》中寫道：  
   **「我們淹沒在資料之中，卻飢渴於知識。」**  
   因此資料集很可能已經存在，只是散落在不同的格式或來源：  
   - 外部或內部  
   - 結構化或非結構化  
   - 靜態或串流  
   - 客觀或主觀  
   你不需要重新發明輪子，只需要知道去哪裡找。  
   下一步我們會將「髒資料」變成「乾淨資料」。

3. **準備可使用的資料。**  
   這個步驟通常稱為 **資料整理（Data Wrangling）**：  
   將「野生」資料變成「可管理」資料。  
   資料整理包含：  
   - 建立資料儲存與處理架構  
   - 建立資料治理與品質標準  
   - 資料擷取（ETL、爬蟲）  
   - 資料清理：找出異常值、缺失值、離群值  

4. **進行探索性分析（EDA）。**  
   所有人都知道：**Garbage In, Garbage Out（垃圾進、垃圾出）**。  
   因此需要使用描述性統計、圖形統計來找出：  
   - 潛在問題  
   - 模式  
   - 分類  
   - 相關性  
   - 比較  
   同時，理解資料型態（類別型 vs 連續型）也很重要，  
   這會影響後續使用的假設檢定或模型類型。

5. **建立模型。**  
   就像描述統計與推論統計一樣，資料模型可以用來：  
   - 總結資料  
   - 預測未來結果  
   演算法不是魔法棒，而是工具。  
   如果你說你需要十字螺絲起子，而對方給你一支一字起子甚至是鐵鎚，  
   最好的情況是不專業，最糟的是無法完成任務。  
   資料建模也是如此：選擇合適的工具非常重要。  
   **你才是工匠，而演算法只是工具。**

6. **驗證並實作模型。**  
   在使用訓練集建立模型後，必須用測試集進行驗證。  
   這是為了避免：  
   - 過度擬合（overfitting）  
   - 使模型只適用於特定資料，而無法泛化到同一資料集的其他子集  

7. **優化與制定策略。**  
   這是「六百萬美元人（The Bionic Man）」的步驟：  
   不斷迭代，讓模型變得更好、更強、更快。  
   身為資料科學家，你應該盡可能將開發作業與系統整合外包或自動化，  
   讓你能專注於：  
   - 洞察  
   - 推薦  
   - 設計  
   當你能清楚包裝你的想法，它們就變成你在組織內的「價值交換貨幣」。

## 3.21 資料清理的四大要素：  
### Correcting（修正）、Completing（補全）、Creating（建立）、Converting（轉換）

在此階段，我們將透過以下四個步驟清理資料：  
1) 修正異常值與離群點  
2) 補全缺失資料  
3) 建立新特徵  
4) 將欄位轉換為適合模型計算的格式  

---

### 1. **Correcting（修正）**  
檢查資料後，目前看起來沒有明顯異常或不可接受的輸入值。  
雖然 *Age* 與 *Fare* 可能包含離群值，但它們皆屬合理範圍，因此我們會等到完成探索性分析（EDA）後，再判斷是否應排除。  

如果遇到完全不合理的數值，如 `age = 800`（應為 80），此時就可立即修正。  
但需注意：**修改原始資料需謹慎，以免破壞模型的真實性。**

---

### 2. **Completing（補全）**  
*Age*、*Cabin*、*Embarked* 欄位存在缺失值。  
缺失資料會讓模型產生偏誤，因此在建模前必須先處理。

處理方式常見兩種：  
- **刪除紀錄**（不建議，除非該筆資料確實不可用）  
- **填補資料（imputation）**（較佳作法）

推薦方法：  
- 類別型資料：填補「眾數（mode）」  
- 數值型資料：填補「平均／中位數／平均＋隨機標準差」  

進階方法可依條件填補，如：  
- 依 Pclass 的平均年齡填補 Age  
- 依 Fare 與 SES 推估 Embarked  

但在採用複雜方法前，應先確定其確實能改善基本模型的效能。

本資料集採用：  
- Age → 以中位數填補  
- Cabin → 缺失太多，整欄捨棄  
- Embarked → 以眾數填補  

（後續模型迭代時可再調整，以觀察是否提高準確率）

---

### 3. **Creating（建立）**  
特徵工程（Feature Engineering）指使用現有資料來創造新特徵，以提升模型能力。

在本資料集中，我們將從姓名欄位中萃取「稱謂 Title」作為新特徵，以檢查其是否影響生還率。

---

### 4. **Converting（轉換）**  
最後是資料格式的處理。  
本資料集中雖無日期或貨幣格式，但有「資料型態」問題。

部分欄位（如 Sex、Embarked）以 *object* 型態匯入，不適合直接進行數學運算，因此需要轉換。

在本資料集中，我們會將名目型欄位轉換成：

- **虛擬變數（dummy variables）**

```python
# 補全：在訓練與測試／驗證資料中填補或刪除缺失值
for dataset in data_cleaner:    
    # 使用中位數填補缺失的年齡
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)

    # 使用眾數填補缺失的登船港口
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)

    # 使用中位數填補缺失的票價
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)
    
# 刪除 Cabin、Ticket 與 PassengerId 欄位
drop_column = ['PassengerId','Cabin', 'Ticket']
data1.drop(drop_column, axis=1, inplace = True)

print(data1.isnull().sum())
print(data_val.isnull().sum())
```

```python
# 建立：為訓練與測試／驗證資料進行特徵工程
for dataset in data_cleaner:    

    # 建立離散型特徵：家庭人數
    dataset['FamilySize'] = dataset ['SibSp'] + dataset['Parch'] + 1

    # 建立是否獨自旅行特徵，預設為 1（獨自）
    dataset['IsAlone'] = 1
    # 若家庭人數 > 1，則為 0（非獨自）
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0

    # 從姓名欄位中萃取稱謂（Title）
    dataset['Title'] = dataset['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]

    # 使用 qcut 將票價分成四個區間
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4)

    # 使用 cut 將年齡分成五個區間
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)


# 再次預覽資料
data1.info()
data_val.info()
data1.sample(10)
```

## 了解機器學習的目的

機器學習的目的是**解決人類的問題**。

1. **監督式學習（Supervised Learning）**  
   模型訓練時會給定正確答案（目標變數）。
2. **非監督式學習（Unsupervised Learning）**  
   訓練資料沒有正確答案。
3. **強化學習（Reinforcement Learning）**  
   模型透過「行為 → 回饋」的方式逐步學習。

本專案使用的是 **監督式學習**，因為我們提供：

## 確認使用的演算法類型：分類（Classification）

在確定使用 **監督式學習＋分類模型** 之後，可以將選擇的演算法範圍縮小到：

### 常見分類演算法（scikit-learn）
* [Ensemble Methods（集成演算法）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.ensemble)
* [Generalized Linear Models（GLM）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model)
* [Naive Bayes（朴素貝氏）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.naive_bayes)
* [Nearest Neighbors（KNN）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.neighbors)
* [Support Vector Machines（SVM）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.svm)
* [Decision Trees（決策樹）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.tree)


```python
from sklearn import model_selection, ensemble, gaussian_process, linear_model, naive_bayes, neighbors, svm, tree, discriminant_analysis

# 機器學習演算法（MLA）清單
MLA = [
    # 集成方法 Ensemble
    ensemble.AdaBoostClassifier(),
    ensemble.BaggingClassifier(),
    ensemble.ExtraTreesClassifier(),
    ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(n_estimators=100),

    # 高斯過程
    gaussian_process.GaussianProcessClassifier(),

    # 一般化線性模型（GLM）
    linear_model.LogisticRegressionCV(max_iter=500),
    linear_model.PassiveAggressiveClassifier(),
    linear_model.RidgeClassifierCV(),
    linear_model.SGDClassifier(),
    linear_model.Perceptron(),

    # 朴素貝氏
    naive_bayes.GaussianNB(),

    # 最近鄰演算法
    neighbors.KNeighborsClassifier(n_neighbors=3),

    # 支援向量機
    svm.SVC(probability=True),
    svm.LinearSVC(),

    # 決策樹
    tree.DecisionTreeClassifier(),
    tree.ExtraTreeClassifier(),

    # 判別分析
    discriminant_analysis.LinearDiscriminantAnalysis(),
    discriminant_analysis.QuadraticDiscriminantAnalysis()
]

# 建立演算法比較表
MLA_columns = ['MLA Name', 'MLA Parameters', 'MLA Train Accuracy', 'MLA Test Accuracy', 'MLA Time']
MLA_compare = pd.DataFrame(columns=MLA_columns)

Target = 'Survived'
y = data1[Target].values.ravel()

row_index = 0
for alg in MLA:
    print(f"Running: {alg.__class__.__name__}")

    MLA_compare.loc[row_index, 'MLA Name'] = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Parameters'] = str(alg.get_params())

    # 進行 10-fold 交叉驗證（包含訓練與測試分數）
    cv_results = model_selection.cross_validate(
        alg,
        data1[data1_x_bin],
        y,
        cv=10,
        return_train_score=True,
        n_jobs=-1
    )

    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
    MLA_compare.loc[row_index, 'MLA Train Accuracy'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy'] = cv_results['test_score'].mean()

    # 在完整資料集上重新訓練模型（交叉驗證之後）
    MLA[row_index] = alg.fit(data1[data1_x_bin], data1[Target])

    row_index += 1

# 排序顯示結果（依測試準確率由高到低）
MLA_compare.sort_values(by=['MLA Test Accuracy'], ascending=False, inplace=True)
print(MLA_compare)
MLA_compare
```

## Data Science 101：如何手動建立模型（打造你自己的分類模型）

我們現在的準確率持續提升，但能更好嗎？

接下來，我們將 **手動建立一棵決策樹模型**，

```python
# 重要提示：此為手工模型，僅用於教學示範
# 修正版（2025）語法相容

import random
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
import itertools

# ========== 1. 建立隨機預測（Random_Predict） ==========
data1["Random_Predict"] = 0

for index, row in data1.iterrows():
    if random.random() > 0.5:
        data1.at[index, "Random_Predict"] = 1
    else:
        data1.at[index, "Random_Predict"] = 0

# 評估隨機模型準確率
data1["Random_Score"] = (data1["Random_Predict"] == data1["Survived"]).astype(int)

print("Coin Flip Model Accuracy: {:.2f}%".format(data1["Random_Score"].mean() * 100))
print(
    "Coin Flip Model Accuracy w/SciKit: {:.2f}%".format(
        metrics.accuracy_score(data1["Survived"], data1["Random_Predict"]) * 100
    )
)

# ========== 2. 建立透視表（Pivot Table） ==========
pivot_female = data1[data1.Sex == "female"].groupby(["Sex", "Pclass", "Embarked", "FareBin"])[
    "Survived"
].mean()
print("\n\nSurvival Decision Tree w/Female Node: \n", pivot_female)

pivot_male = data1[data1.Sex == "male"].groupby(["Sex", "Title"])[
    "Survived"
].mean()
print("\n\nSurvival Decision Tree w/Male Node: \n", pivot_male)

# ========== 3. 手刻決策樹模型（簡易版） ==========

def mytree(df):
    Model = pd.DataFrame(index=df.index)
    Model["Predict"] = 0

    male_title = ["Master", "Sir"]

    for idx, row in df.iterrows():
        # 預設：死亡
        Model.at[idx, "Predict"] = 0

        # 女性多數生還 → 預測 1
        if row["Sex"] == "female":
            Model.at[idx, "Predict"] = 1

        # 女性例外條件（改善準確率用）
        if (
            row["Sex"] == "female"
            and row["Pclass"] == 3
            and row["Embarked"] == "C"
            and 8 <= row["Fare"] <= 15
        ):
            Model.at[idx, "Predict"] = 0

        if (
            row["Sex"] == "female"
            and row["Pclass"] == 3
            and row["Embarked"] == "S"
            and row["Fare"] > 8
        ):
            Model.at[idx, "Predict"] = 0

        # 男性特殊稱謂（Master, Sir）→ 生還
        if row["Sex"] == "male" and row["Title"] in male_title:
            Model.at[idx, "Predict"] = 1

    return Model


# 執行手刻模型
Tree_Predict = mytree(data1)

print(
    "\n\nDecision Tree Model Accuracy: {:.2f}%\n".format(
        metrics.accuracy_score(data1["Survived"], Tree_Predict["Predict"]) * 100
    )
)

print(metrics.classification_report(data1["Survived"], Tree_Predict["Predict"]))

# ========== 4. 混淆矩陣繪圖 ==========

def plot_confusion_matrix(cm, classes, normalize=False, title="Confusion matrix", cmap=plt.cm.Blues):

    if normalize:
        cm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")

    print(cm)

    plt.imshow(cm, interpolation="nearest", cmap=cmap)
    plt.title(title)
    plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45)
    plt.yticks(ticks, classes)

    fmt = ".2f" if normalize else "d"
    threshold = cm.max() / 2.0

    # 在格子中顯示數值
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(
            j,
            i,
            format(cm[i, j], fmt),
            horizontalalignment="center",
            color="white" if cm[i, j] > threshold else "black",
        )

    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")


# 產生混淆矩陣
cm = metrics.confusion_matrix(data1["Survived"], Tree_Predict["Predict"])

class_names = ["Dead", "Survived"]

plt.figure()
plot_confusion_matrix(cm, class_names, title="Confusion matrix")

plt.figure()
plot_confusion_matrix(cm, class_names, normalize=True, title="Normalized confusion matrix")
plt.show()
```

## 5.11 使用交叉驗證（Cross-Validation, CV）評估模型效能

在步驟 5 中，我們使用了  
[sklearn cross_validate](http://scikit-learn.org/stable/modules/cross_validation.html#multimetric-cross-validation)  
來訓練、測試並評估模型效能。

記住一件事：  
為了評估模型是否具有「真正的預測能力」，我們**一定要**把資料分成：

- **訓練資料（train）**：用來建立模型  
- **測試資料（test）**：用來評估模型

否則模型會 **overfitting（過度擬合）**：  
模型只會「記住」訓練資料的答案，而不是學到可泛化的規則。  
這就像作業和小考都偷看答案拿 100 分，但到了正式考試卻完全不會寫。

---

## 為什麼要用 Cross-Validation？

當資料集很小時，  
強行切割成訓練資料與測試資料，  
會使模型損失大量可用訊息。

**交叉驗證（CV）** 的做法是：

1. 把資料切成 k 份  
2. 每次取其中 1 份當測試資料，其餘 k-1 份當訓練資料  
3. 重複 k 次  
4. 取得所有測試結果的「平均值」當最終效能

這樣可以：

- **充分利用所有資料**
- **避免單次切割造成的偏差**
- **得到更可靠的效能估計**

缺點是：  
計算量增加，需要更多電腦資源。

## 決策樹（Decision Tree）優點  
資料來源：sklearn 官方文件  
<http://scikit-learn.org/stable/modules/tree.html#classification>

> **決策樹的優點：**
* 易於理解與解釋，並可視覺化。
* 幾乎不需要資料前處理（不需正規化、不需 dummy 變數，雖然不能處理缺失值）。
* 預測成本低（時間複雜度與訓練資料筆數呈對數關係）。
* 能同時處理數值與類別資料（其他演算法通常只擅長某一種）。
* 能處理多輸出問題（multi-output）。
* 屬於「白箱模型」：邏輯透明，容易解釋（相較於黑箱如神經網路）。
* 可用統計方法評估其可靠性。
* 即使資料不完全符合演算法假設，也通常能維持良好表現。

---

## 決策樹（Decision Tree）缺點

> **決策樹的缺點：**
* 容易產生過度複雜的樹，導致 **overfitting（過度擬合）**。  
  → 必須使用剪枝、限制最大深度、設定葉節點最小樣本數等方法（sklearn 不支援傳統剪枝）。
* 對資料非常敏感，小變動可能生成完全不同的樹。  
  → 可透過集成方法（ensemble）改善。
* 找到最佳決策樹是 NP-complete 問題，因此實務演算法多為 greedy heuristics，  
  不能保證得到全域最佳的樹。
* 某些模式難以用樹表達（如 XOR、Parity、Multiplexer 等）。
* 若資料不平衡，決策樹容易產生偏差。  
  → 建議在訓練前先平衡資料。
